## Pair Plots:

```
Plots between all pairs of numerical features.

Syntax:
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset(file_path)
sns.pairplot(df)
plt.show()
```

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset('tips')
df.head()
````

````python
sns.pairplot(df)
plt.show()
````

### Pair plots for different categories:

````python
sns.pairplot(df, hue='sex')
plt.show()
````

````python
sns.pairplot(df, hue='day')
plt.show()
````

### Pair plots between selected columns:

````python
sns.pairplot(df, vars=['total_bill', 'tip'], hue='sex')
plt.show()
````
