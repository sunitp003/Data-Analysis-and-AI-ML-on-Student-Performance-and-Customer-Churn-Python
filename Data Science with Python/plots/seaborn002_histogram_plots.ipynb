````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset("penguins")
df.head()
````

````python
df.describe()
````

````python
sns.histplot(df['bill_length_mm'])
plt.show()
````

### Setting bins:

````python
sns.histplot(df['flipper_length_mm'])
plt.show()
````

````python
sns.histplot(df['flipper_length_mm'], bins=[160,180,200,220,240])
plt.show()
````

### KDE parameter: (Kernel Density Estimator)

````python
sns.histplot(df['flipper_length_mm'], kde=True)
plt.show()
````

````python
data = np.random.randn(10000)
sns.histplot(data, kde=True)
plt.show()
````
