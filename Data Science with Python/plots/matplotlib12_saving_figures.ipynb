## Saving figures using matplotlib:
```
Syntax:
x = []
y = []
plt.plot(x,y)
plt.savefig(file_name)
plt.show()
```

````python
import matplotlib.pyplot as plt
````

````python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')

plt.show()
````

#### We can save above figure as image by dragging it into image editor.

````python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig("line-graph") # default filetype: png
plt.show()
````

### dpi parameter:

````python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig("line-graph1", dpi=800, ) # default filetype: png
plt.show()
````

### facecolor: used for background color in the image

````python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig("line-graph2", dpi=800, facecolor='g')
plt.show()
````

### format parameter:

````python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig("line-graph3.pdf", dpi=800, facecolor='c')
plt.show()
````

### Transparency: It helps us to remove white background of the graph

````python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig("line-graph4.jpg", dpi=800, facecolor='c',
           transparent=True)
plt.show()
````

### bbox_inches parameter: to reduce extra space around the graph in the image.

bbox_inches = 'tight' or None(default)

````python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig("line-graph5.jpg", dpi=800, facecolor='c',
           transparent=True, bbox_inches='tight')
plt.show()
````
