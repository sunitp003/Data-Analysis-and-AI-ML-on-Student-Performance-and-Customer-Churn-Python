## Scatter Plot:
```
Syntax:
x = []
y = []
plt.scatter(x,y)
plt.show()
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

### Basic plot

````python
x = [1,2.5,3,4,5,6,7]
y = [2,3,1,3.5,5,2,6]

plt.scatter(x,y)
plt.show()
````

### Adding decorations:

````python
day = [1,2,3,4,5,6,7]
num = [2,3,1,3,5,2,6]

plt.title('Scatter plot', fontsize=15)
plt.xlabel('Day', fontsize=13)
plt.ylabel('Number of Patients', fontsize=13)
plt.title("Clinic records", fontsize=16)
plt.scatter(day,num)
plt.show()
````

### Setting colors:

````python
day = [1,2,3,4,5,6,7]
num = [2,3,1,3,5,2,6]

plt.title('Scatter plot', fontsize=15)
plt.xlabel('Day', fontsize=13)
plt.ylabel('Number', fontsize=13)
plt.scatter(day,num, color='r')
plt.show()
````

### Changing dot shape and size: [Bubble chcart]

````python
from random import random
day = [1,2,3,4,5,6,7]
num = [2,3,1,3,5,2,6]
c = [(random(),random(),random(),1) for i in range(len(day))]
sizes = np.array(num)*100

plt.title('Scatter plot', fontsize=15)
plt.xlabel('Day', fontsize=13)
plt.ylabel('Number', fontsize=13)
plt.scatter(day,num, c=c, s=sizes, alpha=0.7)
plt.show()
````

## Advanced Formating:

### Edge colors:

````python
from random import random
day = [1,2,3,4,5,6,7]
num = [2,3,1,3,5,2,6]
c = [(random(),random(),random(),1) for i in range(len(day))]
sizes = np.array(num)*100

plt.title('Scatter plot', fontsize=15)
plt.xlabel('Day', fontsize=13)
plt.ylabel('Number', fontsize=13)
plt.scatter(day,num, c=c, s=sizes, alpha=0.7, 
            marker='*', edgecolors='r',
           linewidths=2) # linewidth for edge thickness
plt.show()
````

### colorbar() to see the colors domain:

````python
from random import random
day = [1,2,3,4,5,6,7]
num = [2,3,1,3,5,2,6]
c = [(random(),random(),random(),1) for i in range(len(day))]
sizes = np.array(num)*100

plt.title('Scatter plot', fontsize=15)
plt.xlabel('Day', fontsize=13)
plt.ylabel('Number', fontsize=13)
plt.scatter(day,num, c=c, s=sizes, alpha=0.7, 
            marker='>', edgecolors='r',
           linewidths=2) # linewidth for edge thickness
plt.colorbar()
plt.show()
````

### cmap to choose color templates:

cmap values: https://matplotlib.org/stable/tutorials/colors/colormaps.html

````python
from random import randint,random
day = [1,2,3,4,5,6,7]
num = [2,3,1,3,5,2,6]
c = [(randint(0,100)) for i in range(len(day))]
sizes = np.array(num)*100

plt.title('Scatter plot', fontsize=15)
plt.xlabel('Day', fontsize=13)
plt.ylabel('Number', fontsize=13)
plt.scatter(day,num, c=c, s=sizes, 
            marker='^', edgecolors='r',
           linewidths=1, cmap='Accent')
# t = plt.colorbar()
# t.set_label('Color bar') # label for color bar
# One liner:
plt.colorbar().set_label('Color Bar')
plt.show()
````

### Multiple scatter plots:

````python
from random import randint,random
day = [1,2,3,4,5,6,7]
num = [(randint(1,10)) for i in range(len(day))]
num1 = [(randint(1,10)) for i in range(len(day))]
c = [(randint(0,100)) for i in range(len(day))]
c1 = [(randint(0,100)) for i in range(len(day))]
sizes = np.array(num)*100

plt.title('Scatter plot', fontsize=15)
plt.xlabel('Day', fontsize=13)
plt.ylabel('Number', fontsize=13)
plt.scatter(day,num, c='r', s=sizes, marker='^')
plt.scatter(day,num1, c='b', s=sizes, marker='v')
# t = plt.colorbar()
# t.set_label('Color bar') # label for color bar
# One liner:
plt.colorbar().set_label('Color Bar')
plt.show()
````
