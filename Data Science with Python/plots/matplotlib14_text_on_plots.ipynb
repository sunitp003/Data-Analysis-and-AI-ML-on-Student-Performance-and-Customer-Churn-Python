## Text on plots:

* xlabel
* ylabel
* title
* Text: Add text at an arbitrary location 
* Annotate: Add an annotation (with an optional arrow) at an arbitrary position
* Legend

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)
plt.show()
```

### text in plot:

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.text(2.8,5.2,"Local Max")

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.text(2.5,5,"Local Max", fontsize=15, style='italic')

plt.show()
```

### Highlight text with box:

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.text(2,5.5,"Local Max", fontsize=15, style='italic',
        bbox={'facecolor':'c'})

plt.show()
```

### Annotation:

xy: position of arrow pointer

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.text(2,5.5,"Local Max.", fontsize=15, style='italic',
        bbox={'facecolor':'c'})
plt.annotate("Global Min.", xy=(3,4), xytext=(2.5,2))

plt.show()
```

### Annotation with arrow:

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.text(2,5.5,"Local Max.", fontsize=15, style='italic',
        bbox={'facecolor':'c'})
plt.annotate("Global Min.", xy=(2,1), xytext=(2.5,2),
            arrowprops=dict(facecolor='c',shrink=100), fontsize=15)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

plt.text(2,5.5,"Local Max.", fontsize=15, style='italic',
        bbox={'facecolor':'c'})
plt.annotate("Local Min.", xy=(4,2), xytext=(2.5,2),
            arrowprops=dict(facecolor='c',shrink=100), fontsize=15)

plt.show()
```

### Legend:

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)
plt.title('Matplotlib graph', fontsize=20)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)

# plt.text(2,5.5,"Local Max.", fontsize=15, style='italic',
#         bbox={'facecolor':'c'})
plt.annotate("Global Min.", xy=(2,1), xytext=(2.5,2),
            arrowprops=dict(facecolor='c',shrink=100), fontsize=15)

plt.legend(['line-plot'], loc=9, facecolor='y', fontsize=15,
           edgecolor='b', framealpha=0.4, shadow=True)

plt.show()
```
