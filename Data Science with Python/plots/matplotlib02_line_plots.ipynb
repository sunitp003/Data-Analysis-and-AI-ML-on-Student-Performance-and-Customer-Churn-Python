## Line Plot:
```
Syntax:
x = []
y = []
plt.plot(x,y)
plt.show()
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

### Simple line plot:

````python
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
plt.plot(x,y)
plt.show()
````

### Adding labels to the line plot:

````python
plt.xlabel('Language')
plt.ylabel('% popularity')
plt.title('Language Trend')
plt.plot(x,y)
plt.show()
````

### Adjusting label size:

````python
plt.xlabel('Language', fontsize=15) # default:10
plt.ylabel('% popularity', fontsize=15) # default:10
plt.title('Language Trend', fontsize=25) # default:12
plt.plot(x,y)
plt.show()
````

### Adjusting the line thickness:

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
plt.plot(x,y, linewidth=5) # linewidth>=0, default linewidth=1.5
plt.show()
````

### Setting the line color and style:

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
plt.plot(x,y,"go",linewidth=3) # g:green, o:circular marker
plt.show()
````

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
plt.plot(x,y,"r--",linewidth=3) # r:red, --:dashed line
plt.show()
````

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
plt.plot(x,y,color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.show()
````

## Plotting multiple lines:

````python
x = [1, 2, 3]
y = np.array([[1, 2], [3, 4], [5, 6]])
plt.plot(x, y)
plt.show()
````

````python
x = [1, 2, 3]
y = np.array([[1, 2], [3, 4], [5, 6]])
for col in range(y.shape[1]):
    plt.plot(x, y[:, col])
plt.show()
````

````python
x1 = [1,2,3,4,5]
y1 = [2,4,6,8,10]
x2 = [5,4,3,2,1]
y2 = [1,3,5,7,9]
plt.plot(x1, y1, 'r^', x2, y2, 'b-')
plt.show()
````

## Creating functional plots:

````python
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.sin(x)
plt.plot(x,y,color='b')
plt.plot(np.zeros(len(x)),x, color='r')
plt.plot(x,np.zeros(len(x)), color='r')
plt.show()
````

````python
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.cos(x)
plt.plot(x,y,color='b')
plt.plot(np.zeros(len(x)),x, color='r')
plt.plot(x,np.zeros(len(x)), color='r')
plt.show()
````
