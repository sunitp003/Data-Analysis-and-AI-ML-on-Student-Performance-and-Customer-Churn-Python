````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset('penguins')
df.describe()
````

````python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df)
plt.show()
````

### hue parameter:

````python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df,
               hue='species')
plt.show()
````

### style change:
use different symbols for different hue-classes

````python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df,
               hue='species', style='sex')
plt.show()
````

### size:

````python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df,
               hue='sex', size='sex', style='sex', sizes=(100,30))
plt.show()
````
