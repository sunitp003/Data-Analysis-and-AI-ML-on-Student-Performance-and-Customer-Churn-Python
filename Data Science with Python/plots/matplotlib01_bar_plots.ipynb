## Bar Plot:
```
Syntax:
x = []
y = []
plt.bar(x,y)
plt.show()
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

### Simple bar plot:

````python
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
plt.bar(x,y)
plt.show()
````

### Adding labels to the bar plot:

````python
plt.xlabel('Language')
plt.ylabel('% popularity')
plt.title('Language Trend')
plt.bar(x,y)
plt.show()
````

### Adjusting label size:

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
plt.bar(x,y)
plt.show()
````

### Adjusting the bar width:

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
plt.bar(x,y, width=0.4) # default width=0.8, width:[0,1]
plt.show()
````

### Setting the bar color(s):

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
plt.bar(x,y, width=0.4, color='k') # colors: b:blue, r:red, g:green, c:cyan, m:magenta, k:black, so on..
plt.show()
````

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
colors = ['y','b', 'm', 'g']
plt.bar(x,y, width=0.4, color=colors)
plt.show()
````

## Advanced Formating:

### Adjusting the bar positions:

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
colors = ['y','b','g','m']
plt.bar(x,y, width=0.4, color=colors, align='edge') # edge/center(default)
plt.show()
````

### Adding edges to the bars:

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
colors = ['y','b','g','m']
plt.bar(x,y, width=0.4, color=colors, 
        edgecolor='red', linewidth=3)
plt.show()
````

```
linestyle or ls	[solid | dashed, dashdot, dotted | (offset, on-off-dash-seq) | '-' | '--' | '-.' | ':' | 'None' | ' ' | '']
```

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=25)
colors = ['y','b','g','m']
plt.bar(x,y, width=0.4, color=colors, 
        edgecolor='r', linewidth=5, linestyle='-.') # '-', '--', ':', '-.'
plt.show()
````

### Color darkness/transperency variation by 'alpha' parameter

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=20)
colors = ['y','b','g','m']
plt.bar(x,y, width=0.4, color=colors, 
        edgecolor='r', linewidth=5, 
        linestyle=':', alpha=0.4)
plt.show()
````

### Labeling bars via legend:

````python
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=20)
colors = ['y','b','g','m']
plt.bar(x,y, width=0.4, color=colors, 
        edgecolor='r', linewidth=5, 
        linestyle='-', alpha=0.4, label='LEAP')
plt.legend()
plt.show()
````

### Multiple bar graphs:

#### Overlapping bars:

````python
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
z = [20, 30, 40, 50]
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity/Complexity', fontsize=15)
plt.title('Language Trend', fontsize=20)
plt.bar(x,y, width=0.4, color='r', label='Popularity')
plt.bar(x,z, width=0.4, color='b', label='Complexity')
plt.legend()
plt.show()
````

#### Side-by-side bars:

````python
import numpy as np
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
z = [20, 30, 40, 50]
width=0.4
p=np.arange(len(x))
p1=np.arange(len(x))+width
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=20)
plt.bar(p,y, width=0.4, color='r', label='Popularity')
plt.bar(p1,z, width=0.4, color='b', label='Popularity1')
plt.legend()
plt.show()
````

````python
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
z = [20, 30, 40, 50]
width=0.2
p=np.arange(len(x))
p1=np.arange(len(x))+width
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=20)
plt.bar(p,y, width=width, color='r', label='Popularity')
plt.bar(p1,z, width=width, color='b', label='Popularity1')
plt.legend()
plt.show()
````

#### Replacing x-labels with xticks:

````python
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
z = [20, 30, 40, 50]
width=0.2
p=np.arange(len(x)) # [0,1,2,3]
p1=np.arange(len(x))+width #[0.2, 1.2, 2.2, 3.2]
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=20)
plt.bar(p,y, width=width, color='r', label='Popularity')
plt.bar(p1,z, width=width, color='b', label='Popularity1')

plt.xticks(p+width/2, x, rotation=45) # [0.1, 1.1, 2.1, 3.1]
plt.legend()
plt.show()
````

### Horizontal bar chart with barh:

````python
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
z = [20, 30, 40, 50]
# width=0.2
# p=np.arange(len(x))
# p1=np.arange(len(x))+width
plt.ylabel('Language', fontsize=15)
plt.xlabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=20)
plt.barh(x,y, color='r', label='Popularity')
plt.barh(x,z, color='b', label='Popularity1')
plt.legend()
plt.show()
````

````python
x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
z = [20, 30, 40, 50]
width=0.2
p=np.arange(len(x))
p1=np.arange(len(x))+width
plt.xlabel('% popularity', fontsize=15)
plt.ylabel('Language', fontsize=15)

plt.title('Language Trend', fontsize=20)
plt.barh(p,y, width, color='r', label='Popularity')
plt.barh(p1,z, width, color='b', label='Popularity1')

plt.yticks(p+width/2, x, rotation=45)
plt.legend()
plt.show()
````

## Final colorful bar charts:

You can change the color of bars in RGB way of making colors. You have to to provide an amount of red, green, blue, and the transparency value to the color argument and it returns a color.

CMYK color format. :: Cyan, Magenta, Yellow, Key

````python
from random import random

x = ["python", "c", "c++", "java"]
y = [85, 70, 62, 82]
z = [20, 30, 40, 50]
c = [(random(),random(),random(),1) for i in range(len(x))]
c1 = [(random(),random(),random(),1) for i in range(len(x))]

# c = ['r', 'b', 'g', 'c']
width=0.2
p=np.arange(len(x))
p1=np.arange(len(x))+width
plt.xlabel('Language', fontsize=15)
plt.ylabel('% popularity', fontsize=15)
plt.title('Language Trend', fontsize=20)
plt.bar(p,y, width=width, color=c, label='Popularity')
plt.bar(p1,z, width=width, color=c1, label='Popularity1')

plt.xticks(p+width/2, x, rotation=45)
plt.legend()
plt.show()
````
