
````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset('tips')
df.head()
````

### Single graph

````python
fg = sns.FacetGrid(df)
fg.map(plt.scatter,'total_bill','tip')
plt.show()
````

### Seperate graphs for different categories:

````python
fg = sns.FacetGrid(df, col='sex')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
````

````python
fg = sns.FacetGrid(df, col='size')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
````

### hue with legend:

````python
fg = sns.FacetGrid(df, col='day', hue='smoker')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
````

````python
fg = sns.FacetGrid(df, col='day', hue='smoker')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
````

````python
fg = sns.FacetGrid(df, col='day', hue='sex')
fg.map(plt.bar,'tip','total_bill').add_legend()
plt.show()
````

````python
fg = sns.FacetGrid(df, col='day', hue='sex')
fg.map(sns.histplot,'total_bill').add_legend()
plt.show()
````

````python
fg = sns.FacetGrid(df, col='day', hue='sex')
fg.map(sns.histplot,'total_bill',kde=True).add_legend()
plt.show()
````

### Coloring and palette:

````python
fg = sns.FacetGrid(df, col='sex', hue='day', palette='Accent')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
````

````python
fg = sns.FacetGrid(df, col='sex', hue='day', palette='Accent')
fg.map(plt.scatter,'total_bill','tip', edgecolor='r').add_legend()
plt.show()
````
