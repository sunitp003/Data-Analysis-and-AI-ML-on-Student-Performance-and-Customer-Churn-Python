````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

### Matplotlib plot:

````python
x = [1,2,3,4,5,6,7]
y = [2,3,4,1,5,6,7]

plt.plot(x, y)
plt.show()
````

### Seaborn plot:

````python
id = [1,2,3,4,5,6,7]
count = [2,3,4,1,5,6,7]

df = pd.DataFrame({"id":id, "count":count})

sns.lineplot(x="id", y="count", data=df)
plt.show()
````

````python
id = [1,2,3,4,5,6,7]
count = [2,3,4,1,5,6,7]

sns.lineplot(x=id, y=count)
plt.show()
````

### Markers:

````python
df = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=df, hue="sex", style="sex", palette='cool',
            markers=['o','^'])
plt.show()

# to check the number of unique values in a column
len(df["sex"].dropna().unique())
````

### Replace dashed line with solid line:

````python
df = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=df, hue="sex", style="sex", palette='cool',
            markers=['o','^'], dashes=False)

# markers are not displayed if the style is not added
# sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
#              data=df, hue="sex", palette='cool',
#             markers=['o','^'])
plt.show()
````

### Manage legend:

legend parameter values: 'auto', 'brief', 'full' or a boolean

````python
df = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=df, hue="sex", style="sex", palette='cool',
            markers=['o','^'], dashes=False, legend=False)
plt.show()
````

````python
df = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=df, hue="sex", style="sex", palette='cool',
            markers=['o','^'], dashes=False, legend="brief")
plt.show()
````

### Title, grid, etc:

````python
y1 = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=y1, hue="sex", style="sex", palette='cool',
            markers=['o','^'], dashes=False, legend="full")
plt.grid()
plt.title("Line graph example")
plt.show()
````
